```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('Dataset.csv')
data.head()
```

```python
# Import necessary libraries
import pandas as pd
import plotly.express as px
import plotly.offline as pyo

# Convert 'Order Date' to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d/%m/%Y')

# Calculate total sales by order date
total_sales = data.groupby('Order Date')['Sales'].sum().reset_index()

# Create an interactive plot using plotly
fig = px.line(total_sales, x='Order Date', y='Sales', title='Total Sales Over Time', labels={'Sales': 'Total Sales'})

# Initialize plotly offline mode
pyo.init_notebook_mode(connected=True)

# Show the plot
pyo.iplot(fig)
```

```python


# Assuming `data` is already loaded and contains 'Order Date', 'Sales', and 'Product Name'

# Convert 'Order Date' to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d/%m/%Y')

# Total Sales
total_sales = data['Sales'].sum()

# Sales Trends Over Time
# Extract year and month from Order Date
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month

# Group by year and month, summing up sales
monthly_sales = data.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
monthly_sales['Year-Month'] = monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str)

# Plot Sales Trends Over Time using Plotly
fig1 = px.line(monthly_sales, x='Year-Month', y='Sales', markers=True, title='Monthly Sales Trends',
               labels={'Sales': 'Total Sales', 'Year-Month': 'Year-Month'})
fig1.update_layout(xaxis_title='Year-Month', yaxis_title='Total Sales')
fig1.show()

# Best-Selling Products
best_selling_products = data.groupby('Product Name')['Sales'].sum().nlargest(10).reset_index()

# Plot Best-Selling Products using Plotly
fig2 = px.bar(best_selling_products, x='Product Name', y='Sales', title='Top 10 Best-Selling Products',
              labels={'Sales': 'Total Sales', 'Product Name': 'Product Name'})
fig2.update_layout(xaxis_title='Product Name', yaxis_title='Total Sales', xaxis_tickangle=-45)
fig2.show()

# Output Total Sales
print("Total Sales: $", total_sales)
```
